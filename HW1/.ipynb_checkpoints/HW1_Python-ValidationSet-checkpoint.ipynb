```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn import tree
from sklearn.metrics import confusion_matrix
mydata = pd.read_csv("HW1_Data.csv")
```

```python
#Take a look at the data
mydata.head()
```

```python
#We see that churndep is the target variable so split the set into X and Y
X = mydata.values[:, 0:11]
Y = mydata.values[:, 11]

#This first split takes out the test set from the dataset 

X_train, X_test, y_train, y_test = train_test_split( X, Y, random_state = 0)

#This second split takes X_train and splits it into training and validation sets. 
X_train, X_val, Y_train, Y_val = train_test_split(X_train, y_train, test_size = .2, random_state = 0)

#The model should be created using the training set
clf1 = DecisionTreeClassifier(criterion = "entropy", random_state = 0,
                                presort = True, min_samples_split = 2, min_impurity_decrease = 0.0009)
clf1.fit(X_train, Y_train)
```

```python
#This chunk works to build a visual representation of the decision tree. 

#Note: This code was adapted from code that we did in class.
from sklearn.tree import export_graphviz
import graphviz
from IPython.display import Image

X_train = pd.DataFrame(X_train)
print(list(mydata.columns.values))
export_graphviz(
        clf1,
        feature_names = ['revenue', 'outcalls', 'incalls', 'months', 'eqpdays', 'webcap', 'marryyes', 'travel', 'pcown', 'creditcd', 'retcalls',],
        rounded=True,
        filled=True,
        out_file= "HW1_tree.dot"
    )

Image('tree200.png')
```

```python
# Here we use the validation dataset to give us an accuracy. Using this accuracy as feedback, we optimize the parameters in the decision tree call. 
y_pred_val = clf1.predict(X_val)
y_pred_val

print("Model Accuracy with Validation: " + str(accuracy_score(Y_val,y_pred_val)*100))

class_outcomes = [0,1]
```

```python
#Note: This code was adapted from class notes and applied to my decision tree. 
from sklearn.metrics import confusion_matrix
import itertools
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

y_pred = clf1.predict(X_test)
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix'):

    plt.imshow(cm, interpolation='nearest', cmap='Pastel1')
    plt.title(title)
    #plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'

    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color= "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes= class_outcomes,
                      title='Confusion Matrix')
```

```python
print("Model Metrics:")
print()
print("Accuracy: " + str(accuracy_score(y_test,y_pred)*100))
print("Positive Precision: " + str(precision_score(y_test, y_pred)))
print("Negative Precision: " + str(precision_score(y_test, y_pred, pos_label = 0)))
print("Positive Recall: " + str(recall_score(y_test, y_pred)))
print("Negative Recall: " + str(recall_score(y_test, y_pred, pos_label = 0)))
print("Positive F-Measure: " + str(f1_score(y_test, y_pred)))
print("Negative F-Measure: " + str(f1_score(y_test, y_pred, pos_label = 0 )))
```
